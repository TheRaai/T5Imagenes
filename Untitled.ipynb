```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
```

```python
img = cv2.imread('rombo.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_norm = cv2.normalize(gray.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
noise = np.random.random(gray.shape)*0.3
output = gray_norm + noise
```

```python
plt.figure(figsize=(10,8))
plt.imshow(gray_norm,cmap='gray')
plt.show()
plt.figure(figsize=(10,8))
plt.imshow(output,cmap='gray')
plt.show()
```

```python
asd = cv2.bilateralFilter(output,5,10,10)
plt.figure(figsize=(10,8))
plt.imshow(asd, cmap='gray')
plt.show()
```

```python
vector = np.linspace(-0.5, 0.5, output.shape[0])
vector2 = np.linspace(-0.5, 0.5, output.shape[1])
U,V = np.meshgrid(vector2, vector)
a = 5
b = 1
UV = U*a+V*b
H = np.fft.fftshift(np.sinc(np.pi*UV)*np.exp(-1j*np.pi*UV))
W = np.conj(H)/(np.abs(H)**2 + 0.01)
G2 = np.fft.fft2(output)
F = W*G2
iRestored2 = np.real(np.fft.ifft2(F))
plt.figure(figsize=(10,8))
plt.imshow(iRestored2, cmap='gray')
plt.show()
```

```python
x= np.linspace(-127, 128,826)
y= np.linspace(-127, 128,662)
X, Y = np.meshgrid(x,y)
H= np.exp(-0.0003*(X**2+Y**2))
UH = np.fft.fftshift(H)
F = np.fft.fft2(output)
FILT = UH*F
S = np.fft.ifft2(np.fft.fftshift(FILT))
out = abs(S)
plt.figure(figsize=(10,8))
plt.imshow(out,cmap='gray')
plt.show()
```
