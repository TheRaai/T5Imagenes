```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
from skimage import filters,feature
```

```python
img = cv2.imread('rombo.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray_norm = cv2.normalize(gray.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
noise = np.random.random(gray.shape)*0.3
output = gray_norm + noise
```

```python
plt.figure(figsize=(10,8))
plt.imshow(gray_norm,cmap='gray')
plt.show()
plt.figure(figsize=(10,8))
plt.imshow(output,cmap='gray')
plt.show()
```

```python
def filtro_std(A):
    S= np.mean(A.flatten())    
    return S

std_f= ndi.generic_filter(output,filtro_std, [7,7])
plt.figure(figsize=(10,8))
plt.imshow(std_f,cmap='gray')
plt.show()
```

```python
for i in range(1,7):
    print(i)
    canny = feature.canny(std_f,sigma=i)
    cv2.imwrite(f'canny{i}.png',canny*255)
    plt.figure(figsize=(10,8))
    plt.imshow(canny,cmap='gray')
    plt.show()

roberts = filters.roberts(std_f)
plt.figure(figsize=(10,8))
plt.imshow(roberts,cmap='gray')
plt.show()

prewitt = filters.prewitt(std_f)
plt.figure(figsize=(10,8))
plt.imshow(prewitt,cmap='gray')
plt.show()

sato = filters.sato(std_f)
plt.figure(figsize=(10,8))
plt.imshow(sato,cmap='gray')
plt.show()

canny = canny = feature.canny(std_f,sigma=3)
```

```python
canny2 = canny.astype("uint8")
```

```python
kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(25,25))
dila = cv2.dilate(canny2,kernel1)
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(23,23))
erod = cv2.erode(dila,kernel2)
plt.figure(figsize=(10,8))
plt.imshow(erod,cmap='gray')
plt.show()
```

```python
src = erod.copy()
cv2.floodFill(src, None, (10, 10), (255))
plt.figure(figsize=(10,8))
plt.imshow(src,cmap='gray')
plt.show()
```
